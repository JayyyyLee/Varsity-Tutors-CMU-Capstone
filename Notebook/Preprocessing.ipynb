```python
names = os.listdir('/content/drive/MyDrive/transcript/raw')
fil = [i == 40 for i in [len(n) for n in names]]
names = np.array(names)[fil]
```

# 70%

```python
n = []
for i in range(len(raw)):
  if (raw['session_uid'][i]+'.csv') in names:
    d = pd.read_csv('/content/drive/MyDrive/transcript/raw/'+raw['session_uid'][i]+'.csv')
    if (d['Utterance end time (milliseconds)'].iloc[-1]/60000/raw['Tutoring Sessions Duration (minutes)'][i]) >=0.7:
      n.append(raw['session_uid'][i]+'.csv')
```

# student > tutor

```python
nn = []
for i in n:
  data = pd.read_csv('/content/drive/MyDrive/transcript/raw/'+i)
  if not (Counter(data['Speaker'])['student']/Counter(data['Speaker'])['tutor']>2):
    nn.append(i)
```
