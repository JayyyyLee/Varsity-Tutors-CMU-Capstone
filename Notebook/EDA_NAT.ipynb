```python
col = ['who_needs_help',  ' grade_level', ' motivation_for_tutoring', ' struggling_reason']
```

```python
l = []
for i in raw.leadform_initial_request_info:
  r = {}
  if len(str(i).split('|')) != 0:
    if len(str(i).split('|')[0].split(':'))!= 1:
      d = ({j.split(':')[0]: j.split(':')[1] for j in str(i).split('|')})
      for k in col:
        if k in d:
          r[k] = d[k]
        else:
          r[k] =None
    else:
      for k in col:
        r[k] =None
  else:
    for k in col:
      r[k] =None
  l.append(r)
expand_features = pd.DataFrame(l)
```

```python
l = []
for i in expand_features[ ' struggling_reason']:
  for j in str(i).split(','):
    l.append(j.lstrip().rstrip())
Counter(l)
```

```python
a = ['None']
a.extend(list(dict(Counter(l)).keys())[1:])
plt.pyplot.pie(list(dict(Counter(l)).values()))
plt.pyplot.legend(a)
plt.pyplot.title('Reason for struggling')
```
